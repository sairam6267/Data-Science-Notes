## Measure of Central Tendency:

__Measure of central tendency__<font color= blue> __describes the typical or central value where the noise is more, based on distribution of values across data set column it tells what is cental value.Central tendency can be measured by Mean,Median and Mode__ </font>

### Mean.

__Mean__ <font color= dark brown> __describes the average value of data set columns.The function of mean is to sum the values of data set columns and divide the sum values by the number of observations.Mean can be denoted by 'mu' symbol but drawback of mean is it allows very high or very low numbers to compute the outcome__</font>


$$\mu = \frac{x_1+x_2+.....+x_n}{N}$$



$$\mu = \sum_{i=0}^{\infty} \frac{X{i}}{N}$$

__Note: Mean is used when no OUTLIERS are present in Dataset columns__

__Example1 :__
* In a class there are 6 students with different heights (5.9,5.8,6.3,6.2,6.1,6.1) now football association of school  want to check the whether the students are eligible to play football or not ?(avg. height should be 6.0 feet)


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import math as m
from scipy.stats import norm
```

```python
Students=pd.DataFrame({"Heights":[6.3,6.0,6.3,6.2,6.1,6.1]})
Students_Avg_height=Students.Heights.mean()
print("Average height of students in class: ",round(Students_Avg_height,2),"feet")
```

```python
sns.distplot(Students.Heights,kde=True,hist=False)
plt.axvline(Students_Avg_height, color='coral', label='Mean')
```

__Conclusion: Students are eligible to play football__

### Median.

__Median__ <font color= dark brown> __describes the middle value of data set column,basically median is another type of average.It function is to sort the values from ascending to descending and retrives the mid value and if count of observations are even or in odd there is specific formula is used to define median__</font>
* if n is even:
$$Median = \frac{\left( \frac{n}{2} \right)^{th}+\left( \frac{n+1}{2} \right)^{th}}{2}$$

* If n is odd:
$$Median = \left( \frac{n+1}{2} \right)^{th}$$

__Note: Median is used when  OUTLIERS are present in Dataset columns__

## Measure of dispersion or spread

__Measure of dispersion__<font color= blue> __: It is used to find the variability from mean and describes the how much the values are spread over the data set column .Dispersion can be described by the  by range,Interquartile Range, Standard deviation,Variance and Mean absolute deviation.__ <font>

### Range.

__Range__<font color= dark brown> __is used to find the difference between maximun and minimum value.Range value describes the how much the range width is spread over data set column.It may be impacted by the outliers in some cases where the column values are not tightly clustered and it takes only minimum  and maximum value of data set column__ <font>

__Inter Quartile Range (IQR)__ <font color= dark brown> __describes the variability of  measure of dispersion and shows  the minimum and maximum value to see the actual spread irrespective of outliers .It contains Quartiles which describes (Q1) the middle value in the first half of the ordered data set,(Q2) is the median value in the set. and (Q3) is the middle value in the second half of the ordered data set. and IQR tells us waht is actual mean value.__ <font>

* Quartile1 :
$$Q_1 = \left( \frac{n+1}{4} \right)^{th}$$

* Quartile3:
$$Q_2 = \left( \frac{3\left(n+1\right)}{4} \right)^{th}$$

* __Inter Quartile  Range__ Q3:  $$IQR=Q_3 - Q_1$$

* lower limit
$$Q1 - 1.5 * IQR$$
* upper limit
$$Q3 + 1.5 * IQR$$ 

```python
salary_range['Salary'].describe()
```

```python
#  sort the values
Salary_range=salary_range.sort_values('Salary')
```

```python
# find the first quartile
Q1=round((len(salary_range['Salary'])+1)/4,0)
Q1_value=Salary_range.iloc[int(Q1),2]
print("first quartile middle value in the first half of the ordered data set: ",Q1)
print("salary Value of Q1: ",Q1_value)
```

```python
# find the thired quartile
Q3=round(3*(len(salary_range['Salary'])+1)/4,0)
Q3_value=Salary_range.iloc[int(Q3),2]
print("Third quartile middle value in the last half of the ordered data set:: ",Q3)
print("salary Value of Q3: ",Q3_value)
```

```python
# Find the median (inter quartile range)
Q2=Q3-Q1
Q2_value=Salary_range.iloc[int(Q2),2]
print("median value in the order data set",Q2)
print("salary Value of Q2: ",Q2_value)
```

```python
sns.distplot(Salary_range['Salary'],hist=False,rug=True)
plt.axvline(Q2_value,color='yellow',label='median')
plt.axvline(Q1_value, color='red')
plt.axvline(Q3_value, color='red')
```

```python
sns.boxplot(Salary_range['Salary'],color="orange")
plt.show
```

__Variance__<font color= dark brown> __It describes how the individual values are varied from the mean or measurement of spread between numbers.It is calculated by the  difference of each observation  from mean squared to the number of observations.__ <font>

 $$\sigma^2=\sum_{i=1}^{N}\, \frac{\left(x_i-\mu\right)^2} {N}$$

__Standard Deviation__<font color= dark brown> __It describes  the average variability of spread from the mean of dataset column.It is squared mean of variance__ <font>

 $$\sigma=\sqrt{\sum_{i=1}^{N}\,\frac{\left(x_i-\mu\right)^2} {N}}$$

```python
mean=sum(Salary_range['Salary'])/len(Salary_range['Salary'])
print("Avg salary :",mean)
```

```python
def std_dev(obs,mean):
    deviations=[]
    for i in obs:
        deviation=(i-mean)**2
        deviations.append(deviation)
        standard_deviation=m.sqrt(sum(deviations)/len(obs))
    return standard_deviation
```

```python
stand_dev=std_dev(Salary_range['Salary'],mean)
print("The average variability from mean salary column is: ",stand_dev)
```

```python
plt.plot(Salary_range['Salary'], norm.pdf(Salary_range['Salary'], mean, stand_dev))
plt.scatter(stand_dev, 0,color="orange")
plt.scatter(mean, 0,color="red")
plt.show()
```
